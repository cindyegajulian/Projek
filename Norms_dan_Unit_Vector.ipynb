2. buat vektor v

```python
# buat vector v dengan elemen (3,4)
v = np.array([3, 4])
v
```

```python
# hitung nilai L2 norm dari vector v
(3 ** 2 + 4 ** 2) ** (1/2)
```

```python
# tampilkan unit vector dari vector v
# Membuat vektor
v = np.array([3, 4])

# Menghitung panjang (norma L2)
norm = np.linalg.norm(v)

# Menghitung unit vector
unit_v = v / norm

print("Unit vector:", unit_v)
```

```python
# hitung panjang unit vector dari vector v
# Membuat vektor
v = np.array([3, 4])

# Menghitung panjang (norma L2)
norm = np.linalg.norm(v)

# Menghitung unit vector
unit_v = v / norm

print("Panjang unit vector:", np.linalg.norm(unit_v))
```
