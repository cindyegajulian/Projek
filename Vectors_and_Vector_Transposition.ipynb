2. lakukan transpose pada vector x dan simpan pada variabel x_t

```python
# tampilkan ukuran dari vector x_t
x_t.shape
```

```python
# lakukan transpose pada vector y dan simpan pada variabel y_t
y = np.array([[2, 7, 27]]) # pada soal sebelumnya atau soal no 1 vektor y belum dibuat
y                        # jadi ini saya buat dulu untuk bisa menjawab pertanyaan transpose dari y

y_t = y.T
y_t
```
